### Image Processing

```python
# TODO: Process only one image
def phi_map(image_list, as_variable=True):
    # Frame Skipping size
    k = len(image_list)
    
    im_tuple = tuple()
    for i in range(k):    
        # Load single image as PIL and convert to Luminance
        im = PIL.Image.fromarray(image_list[i]).convert('L')
        # Resize image
        im = im.resize((84, 84), PIL.Image.ANTIALIAS)
        # Transform to numpy array
        im = np.array(im)
        # Add processed image to tuple
        im_tuple += (im,)
    
    # Return tensor of processed images
    arr = tuple_to_numpy(im_tuple)
    # Convert to Variable
    if as_variable: arr = Variable(torch.from_numpy(arr).float())
    return arr

def tuple_to_numpy(im_tuple):
    # Stack tuple of 2D images as 3D np array
    arr = np.dstack(im_tuple)
    # Move depth axis to first index: (height, width, depth) to (depth, height, width)
    arr = np.moveaxis(arr, -1, 0)
    # Make arr 4D by adding dimension at first index 
    arr = np.expand_dims ( arr, 0 )
    return arr
```

### Model

```python
class DeepQNetwork(nn.Module):
    
    def __init__(self, num_actions):
        super(DeepQNetwork, self).__init__()
        self.conv1 = nn.Sequential(
            nn.Conv2d(4, 32, kernel_size=8, stride=4),
            nn.ReLU()
        )
        self.conv2 = nn.Sequential(
            nn.Conv2d(32, 64, kernel_size=4, stride=2),
            nn.ReLU()
        )
        self.conv3 = nn.Sequential(
            nn.Conv2d(64, 64, kernel_size=3, stride=1),
            nn.ReLU()
        )
        self.hidden = nn.Sequential(
            nn.Linear(64 * 7 * 7, 512),
            nn.ReLU()
        )
        self.out = nn.Sequential(
            nn.Linear(512, num_actions),
            nn.ReLU()
        )
    
    def forward(self, x):
        x = self.conv1(x)
        x = self.conv2(x)
        x = self.conv3(x)
        x = x.view(x.size()[0], -1)
        x = self.hidden(x)
        x = self.out(x)
#         print(x.size())
        return x
```

```python
class ReplayMemory():
    
    def __init__(self, N):
        self.N = N
        self.list = []
        self.next_index = 0
    
    def add(mem):
        '''
        This operation adds a new experience e, replacing the earliest experience if memory is full.
        '''
        # Add new element if memory is not full
        if len(self.list) <= self.next_index:
            self.list.append(mem)
        # Otherwise replace value in position next_index
        else:
            self.list[self.next_index] = mem
        # Update value of next index
        self.index = (self.index + 1) % self.N
        
    def sample():
        return np.random.choice(self.list, MINIBATCH_SIZE)
        
        
class History():
    
    def __init__(self):
        self.max_size = 4
        self.list = []
    
    def add(ex):
        # Add new element if list is not full
        if len(self.list) < self.max_size:
            self.list.append(mem)
            return
        
        # Move existing elements one index to the left
        self.list[:-1] = self.list[1:]
        # Add new value to last index
        self.list[-1] = ex
        
    def get():
        return self.list
        
```

```python
def initial_history(env):
    s = env.reset()[0]
    H = History()
    for _ in range(H.max_size):
        H.add(s)

def e_greedy_action(Q, phi, env, e=.99):
    # Obtain a random value in range [0,1)
    rand = np.random.uniform()
    # With probability e select random action a_t
    if rand < e:
        return env.action_space.sample()
    # Otherwise select a_t = argmax_a Q(phi, a)
    else:
        return(Q(phi).max()[0])
    
def update_target_network(Q):
    return copy.deepcopy(Q)
```
